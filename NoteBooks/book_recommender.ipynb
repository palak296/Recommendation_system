```python
books = pd.read_csv('input/books_1.Best_Books_Ever.csv')
books.head(3)
```

```python
books = books[books['language']=='English']
```

```python
books.dropna(inplace=True,how='all')
```

```python
books = books[['bookId','title','series','author','rating','description','language','genres','publisher','coverImg']]
```

```python
books = books.loc[books['rating'] >= 4.45]
```

```python
books = books.drop_duplicates(subset='bookId', keep="first")
```

```python
books = books.iloc[:3000,:]
```

```python
books.count()
```

```python
books.reset_index(inplace=True)
```

```python
books.head(3)
```

```python
books['author'] = books['author'].str.split(",").str[0]
```

```python
def checking(x):
    if str(x).isnumeric() == True:
        return " Edition-" + x
    else:
        return ""
```

```python
books['title_author'] = books['title'] + books['series'].str[-1].apply(checking)+" By "+ books['author']
```

```python
books.head()
```

```python
books['description'] = books['description'].apply(lambda x:str(x).replace("'","").lower().split())
```

```python
import ast
def convert(text):
    return ast.literal_eval(text)
```

```python
books['genres'] = books['genres'].apply(convert)
```

```python
booksinfo = books.copy()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
books['genres'] = books['genres'].apply(collapse)
books['author'] = books['author'].str.replace(" ","")
books['language'] = books['language'].str.replace(" ","")
books['publisher'] = books['publisher'].str.replace(" ","")
```

```python
def listing(x):
    L = []
    for i in str(x).split(" "):
        L.append(i)
    return L
```

```python
books['language'] = books['language'].apply(listing)
books['author'] = books['author'].apply(listing)
books['publisher'] = books['publisher'].apply(listing)
```

```python
books['tags'] = books['genres'] + books['language'] + books['author'] + books['publisher']
```

```python
new = books.drop(columns=['description','genres','language','author','publisher','bookId','title','series','coverImg', 'rating'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.drop(columns=['index'],inplace=True)
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
new['title_author'].iloc[0]
```

```python
def recommendbook(book):
    index = new[new['title_author'] == book].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:9]:
        print(new.iloc[i[0]]['title_author'])
```

```python
recommendbook('Harry Potter and the Order of the Phoenix Edition-5 By J.K. Rowling')
```
