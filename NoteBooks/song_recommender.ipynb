```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
albums = pd.read_csv('input/spotify_albums.csv')
artists = pd.read_csv('input/spotify_artists.csv')
tracks = pd.read_csv('input/spotify_tracks.csv')
```

```python
albums.head()
```

```python
albums = albums[['track_id','images']]
```

```python
artists.head()
```

```python
artists = artists[['track_id','name','genres']]
```

```python
artists = artists.rename(columns = {'name':'artist_name'})
```

```python
tracks.head()
```

```python
tracks = tracks.rename(columns = {'id':'track_id'})
tracks = tracks.rename(columns = {'name':'song_name'})
```

```python
tracks = tracks[['track_id','acousticness','danceability','duration_ms','energy','instrumentalness','key','liveness', 'loudness', 'lyrics', 'song_name', 'playlist','popularity','speechiness', 'tempo','valence']]
```

```python
tracks.columns
```

```python
songs = tracks.merge(artists,on='track_id')
```

```python
songs = songs.merge(albums,on='track_id')
songs.head()
```

```python
songs = songs.drop_duplicates(subset='track_id', keep="first")
```

```python
songs = songs.loc[songs['popularity'] >= 60.0]
```

```python
songs.reset_index(inplace=True)
```

```python
songsinfo = songs.copy()
```

```python
songs.drop(columns=['track_id','lyrics','images','index'],inplace=True)
```

```python
songs['song_name_artist_name'] = songs['song_name'] + ' By ' + songs['artist_name']
```

```python
songs.head()
```

```python

import ast
def convert(text):
    if ast.literal_eval(text)==[]:
        return ["None"]
    else:
        return ast.literal_eval(text)
```

```python
songs['genres'] = songs['genres'].apply(convert)
```

```python
def collapse(L):
    L = L.replace(" ","").replace(",","")
    return L
```

```python
songs['playlist'] = songs['playlist'].apply(collapse)
```

```python
songs['genres'] = songs.genres.apply(lambda x: " ".join(x))
```

```python
songs.head()
```

```python
songs = songs[['song_name_artist_name','playlist','genres','artist_name']]
```

```python
songs['tags'] = songs['artist_name'] + songs['genres'] + songs['playlist']
```

```python
songs.head()
```

```python
new = songs[['song_name_artist_name','tags']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

songsimilarity = cosine_similarity(vector)
```

```python
new.head()
```

```python
new[new['song_name_artist_name'] == 'Bring It Back By Wild Boyz!'].index[0]
```

```python
def recommendsong(song):
    index = new[new['song_name_artist_name'] == song].index[0]
    distances = sorted(list(enumerate(songsimilarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]]['song_name_artist_name'])
```

```python
recommendsong('Bowl For Two By The Expendables')
```

```python
songsinfo['song_name_artist_name'] = songsinfo['song_name'] + ' By ' + songsinfo['artist_name']
```

```python
import ast
def convertextra(text):
    if ast.literal_eval(text)==[]:
        return ["None"]
    else:
        return ast.literal_eval(text)
```

```python
songsinfo['genres'] = songsinfo['genres'].apply(convertextra)
```

```python
import ast

def urlextract(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['url']) 
    if len(L)>0:
        return L[0]
    else:
        return ""
```

```python
songsinfo['images'] = songsinfo.images.apply(urlextract)
```

```python
songsinfo.head()
```

```python
songsinfo = songsinfo[['song_name_artist_name','song_name','artist_name','genres','playlist','lyrics','images','popularity']]
```

```python
songsinfo.head()
```

```python
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq

arrays = [
    pa.array(col)  # Create one arrow array per column
    for col in songsimilarity
]

table = pa.Table.from_arrays(
    arrays,
    names=[str(i) for i in range(len(arrays))] # give names to each columns
)
# Save it:
pq.write_table(table, 'songparquet/song_similarity.parquet')
```
